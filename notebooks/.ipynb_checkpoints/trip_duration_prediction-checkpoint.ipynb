```python
import pandas as pd
```

```python
import pickle
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

from sklearn.metrics import mean_squared_error
```

```python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']
```

```python
def read_df(fname):
    df = pd.read_parquet(fname)
    
    df['duration'] = pd.to_datetime(df.lpep_dropoff_datetime) - pd.to_datetime(df.lpep_pickup_datetime)
    df.duration = df.duration.apply(lambda td: td.total_seconds()/60)
    df = df[(df.duration >= 1) & (df.duration <= 60)]
    categorical = ['PULocationID', 'DOLocationID']
    numerical = ['trip_distance']
    df[categorical] = df[categorical].astype('str')
    
    return df
```

```python
df = read_df('data/green_tripdata_2022-01.parquet')
```

```python
df_train = df.copy()
df_valid = read_df('data/green_tripdata_2022-02.parquet')
```

```python
sns.distplot(df.duration)
```

```python
len(df_train), len(df_valid)
```

```python
train_dicts = df[categorical + numerical].to_dict(orient='records')
dv = DictVectorizer()

X_train = dv.fit_transform(train_dicts)
y_train = df['duration'].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_train)

mean_squared_error(y_train, y_pred, squared=False)
```

```python
sns.distplot(y_pred, label='predictions')
sns.distplot(y_train, label='ground truth')
plt.legend()
```

```python
df_train['PU_DO'] = df_train['PULocationID'] + '_' + df_train['DOLocationID']
df_valid['PU_DO'] = df_valid['PULocationID'] + '_' + df_valid['DOLocationID']
```

```python
categorical = ['PU_DO'] #PULocationID', 'DOLocationID']
numerical = ['trip_distance']

dv = DictVectorizer()

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_valid[categorical + numerical].to_dict(orient='records')
X_valid = dv.transform(val_dicts)
```

```python
y_train = df_train['duration'].values
y_valid = df_valid['duration'].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_valid)

mean_squared_error(y_valid, y_pred, squared=False)
```

```python
with open('models/lr.bin', 'wb') as f_out:
    pickle.dump((dv, lr), f_out)
```

```python
sns.distplot(y_pred, label='predictions')
sns.distplot(y_valid, label='ground truth')
plt.legend()
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

y_pred = lasso.predict(X_valid)

mean_squared_error(y_valid, y_pred, squared=False)
```

```python
with open('models/lasso.bin', 'wb') as f_out:
    pickle.dump((dv, lasso), f_out)
```

```python
rid = Ridge(alpha=0.1)
rid.fit(X_train, y_train)

y_pred = rid.predict(X_valid)

mean_squared_error(y_valid, y_pred, squared=False)
```

```python
with open('models/ridge.bin', 'wb') as f_out:
    pickle.dump((dv, rid), f_out)
```
